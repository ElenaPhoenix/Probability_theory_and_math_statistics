## Задание 1

```python
import numpy as np
import matplotlib.pyplot as plt


X1 = np.array([30, 30, 40, 40])
Y1 = np.array([37, 47, 50, 60])
# y = b1 * x + b0, x - влияющий параметр
b1 = (np.mean(X1 * Y1) - np.mean(Y1) * np.mean(X1)) / (np.mean(X1 ** 2) - np.mean(X1) ** 2)
b0 = np.mean(Y1) - b1 * np.mean(X1)
print(f'y = {b1}x + {b0}')
plt.scatter(X1, Y1)
plt.plot(X1, b1 * X1 + b0)
R = np.corrcoef(X1, Y1) ** 2
R[0, 1]
```

## Задача 4

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
x1 = np.array([10,8, 13, 9, 11, 14, 6, 4, 12, 7, 5])
y1 = np.array([8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68 ])
# y = 0.5000909090909076x + 3.0000909090909245
x1 = x1.reshape(-1, 1)
model.fit(x1, y1)
r_sq = model.score(x1, y1)
print(f'Коэф. детерм. {r_sq}')
print(f'intercept {model.intercept_}')
print(f'slope {model.coef_}')
```
